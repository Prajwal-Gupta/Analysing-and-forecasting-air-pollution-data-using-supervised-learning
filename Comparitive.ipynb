```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_roya = pd.read_csv('/content/drive/MyDrive/MiniProject/royapuram_daily.csv')
```

```python
df_chan = pd.read_csv('/content/drive/MyDrive/MiniProject/chandnichowk_daily.csv')
```

```python
def plot_points(cname):
  fig = plt.figure(figsize =(30, 10))
  plt.plot(df_roya[cname], color = 'r', label = 'Royapuram')
  plt.plot(df_chan[cname], color = 'g', label = 'Chandnichowk')
  plt.xlabel(cname)
  plt.title(("Comparision of {} in Royapuram and Chandnichowk in the month of August").format(cname))
    
  plt.legend()
```

```python
# function to print the value of the column on top of it
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], round(y[i],2), ha = 'center')
```

```python
def plot_diff(colname):
  fig = plt.figure(figsize =(30, 10))
  plt.xlabel('Dates')
  plt.ylabel(('Difference between {} values of Royapuram and Chandnichowk').format(colname))
  plt.xticks(rotation = 60)
  addlabels(df_diff['DATE'], df_diff[colname])
  plt.bar(df_diff['DATE'], df_diff[colname])
```

```python
plot_diff('PM2.5')
```

```python
plot_diff('PM10')
```
